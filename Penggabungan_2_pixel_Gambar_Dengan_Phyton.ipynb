```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow # Import cv2_imshow


# Membaca dua citra
citra1 = cv2.imread('al-abrar-mecca-saudi-arabia-hotel-building-1677975b86780d6d2a289f83e292d0cf.jpg')  # Ganti dengan path gambar kamu
citra2 = cv2.imread("ka'bah.jpeg")  # Ganti dengan path gambar kamu

# Ubah ukuran agar citra sama
citra1 = cv2.resize(citra1, (500, 500))  # Ubah ukuran sesuai kebutuhan
citra2 = cv2.resize(citra2, (500, 500))

# Tetapkan nilai alpha
alpha = 0.6  # rasio kontribusi citra pertama
beta = 1.0 - alpha

# Lakukan proses blending
citra_hasil = cv2.addWeighted(citra1, alpha, citra2, beta, 0.0)

# Tampilkan hasil
cv2_imshow(citra_hasil) # Use cv2_imshow instead of cv2.imshow
# cv2.waitKey(0)  # Remove this line as it is not needed in Colab
# cv2.destroyAllWindows() # Remove this line as it is not needed in Colab

# Simpan hasil citra
cv2.imwrite('hasil_penggabungan.jpg', citra_hasil)
```
