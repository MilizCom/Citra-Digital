```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Baca gambar (grayscale)
img = cv2.imread('Gambar WhatsApp 2025-06-17 pukul 16.23.01_e952db4d.jpg', cv2.IMREAD_GRAYSCALE)

# Periksa apakah gambar berhasil dimuat
if img is None:
    print("Gambar tidak ditemukan!")
    exit()

# Kernel high pass versi Anda
kernel_hp_custom = np.array([
    [ 1, -2,  1],
    [-2,  5, -2],
    [ 1, -2,  1]
])

# Terapkan filter
filtered = cv2.filter2D(img, -1, kernel_hp_custom)

# Tampilkan hasil
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.title('Gambar Asli')
plt.imshow(img, cmap='gray')

plt.subplot(1, 2, 2)
plt.title('Hasil High Pass Kernel 2')
plt.imshow(filtered, cmap='gray')
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Baca gambar berwarna (BGR di OpenCV)
# img = cv2.imread('toriyaki-chicken-dark-surface-768x512.jpg')
# img = cv2.imread('1349196.png')
img = cv2.imread('Gambar WhatsApp 2025-06-17 pukul 16.23.01_e952db4d.jpg')

if img is None:
    print("Gambar tidak ditemukan!")
    exit()

# Kernel High Pass Custom
kernel_hp_color = np.array([
    [ 0,  1,  0],
    [ 0,  0,  0],
    [ 0, -1,  0]
])

# Terapkan filter ke tiap channel (B, G, R)
filtered = cv2.filter2D(img, -1, kernel_hp_color)

# Ubah ke RGB untuk ditampilkan dengan matplotlib
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
filtered_rgb = cv2.cvtColor(filtered, cv2.COLOR_BGR2RGB)

# Tampilkan hasil
plt.figure(figsize=(12,6))
plt.subplot(1, 2, 1)
plt.title("Gambar Asli")
plt.imshow(img_rgb)

plt.subplot(1, 2, 2)
plt.title("Hasil High Pass Filter (Warna)")
plt.imshow(filtered_rgb)
plt.show()
```
